# Part 3. Machine Learning Methods To Predict Heart Attack

```python
#import package
from sklearn.model_selection import train_test_split

#split predictor and outcome
predictors = df.drop("HeartDiseaseorAttack",axis=1)
outcome = df["HeartDiseaseorAttack"]

#split train and test sets
X_train,X_test,Y_train,Y_test = train_test_split(predictors,outcome,test_size=0.20,random_state=0)
```

```python
#visualization function of difference between model and actural
def plot_comparison(predicted_data,actural_data,title):
  
  fig, axes = plt.subplots(1, 2, sharex=True, figsize=(10,5))
  fig.suptitle(title)

  sns.countplot(predicted_data,ax=axes[0])
  axes[0].set_title('Predicted')

  sns.countplot(actural_data,ax=axes[1])
  axes[1].set_title('Actual')
```

### Logistic Regression

```python
#import packages
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
#rescale the data to logistic regression
pipe = make_pipeline(StandardScaler(), LogisticRegression())
```

```python
#fit the model find the accurancy score
pipe.fit(X_train, Y_train)
score_lr = pipe.score(X_test,Y_test)
```

```python
#Compare two results
pred = pipe.predict(X_test)
arr = Y_test
plot_comparison(pred,arr,"Logistic regression result comparison")
```

```python
score_lr
```

### SVM

```python
#import package
from sklearn import svm

#fit model find the predicted outcome
sv = svm.LinearSVC()

sv.fit(X_train, Y_train)

Y_pred_svm = sv.predict(X_test)
```

```python
#plot and compare two results
plot_comparison(Y_pred_svm,Y_test,"SVM result comparison")
```

```python
accuracy_score(Y_pred_svm,Y_test)
```

### K Neighbors Classification

```python
#import package
from sklearn.neighbors import KNeighborsClassifier

#fit the classifier and fit the data
knn = KNeighborsClassifier(n_neighbors=21)
knn.fit(X_train,Y_train)
Y_pred_knn=knn.predict(X_test)
#find accurancy score
score_knn = round(accuracy_score(Y_pred_knn,Y_test)*100,2)
```

```python
#plot and compare two results
plot_comparison(Y_pred_knn,Y_test,"KNN result comparison")
```

```python
score_knn
```
